```python
#Dropping unnecessary columns
df1=df.drop(['PassengerId','Name','Cabin','Ticket'],axis=1)
df1
```

```python
#filling missing values of age with mean
a=df1['Age'].mean()
df1['Age'].fillna(a,inplace=True)
```

```python
#Deleting rows of embarked with missing values
df1.dropna(subset=['Embarked'],inplace=True)
df1
```

```python
#encoding method
dummy=pd.get_dummies(df1[['Sex','Embarked']])
dummy
```

```python
#concate with df1
df2 = pd.concat([df1,dummy],axis=1)
df2
```

```python
#drop sex and embarked
df2.drop(['Sex','Embarked'],axis=1,inplace=True)
df2
```

```python
df2.info()
```

```python
categorical = [i for i in df2.columns if df2[i].dtype=='O']
categorical
```

```python
for i in categorical:
  print(df1[i].value_counts())
```

```python
x=df2.iloc[:,1:].values
y=df2.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
mat=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
score
```
